# Jak przekonwertować notebook do PDF żeby ładnie wyglądał?

## Konwersja `notebook` -> `HTML` 

* Otwórz `terminal`
* Upewnij się, że środowisko conda jest aktywne
* Przejdź do folderu z notebookiem
* Uruchom komendę (podmień `09__notebook_as_report.ipynb` oraz `09__notebook_as_report.html` na swoje nazwy plików):
```bash
jupyter nbconvert 09__notebook_as_report.ipynb --to html --no-input --no-prompt --output 09__notebook_as_report.html
```

#### Konwersję na `HTML` można wykonać również z poziomu `Jupyter Notebook`:

````python
!jupyter nbconvert 09__notebook_as_report.ipynb --to html --no-input --no-prompt --output 09__notebook_as_report.html
````

## Konwersja `HTML` -> `PDF`

* otwórz folder, w którym znajduje się plik `HTML`
* kliknij prawym przyciskiem myszy na plik `HTML`
* otwórz plik w przeglądarce
* wydrukuj stronę do pliku `PDF`
